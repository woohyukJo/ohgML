```
import requests
from bs4 import BeautifulSoup
import pandas as pd
jeju_data = requests.get('http://www.jeju.go.kr/jejuculture/show/show.htm?year=2020&month=10')
jeju_data_soup = BeautifulSoup(jeju_data.text,'html.parser')
ex_container = jeju_data_soup.find_all('div',{'class':"ex-container"})
performlist = []
for i in ex_container:
  p_find_all = i.find_all('p')
  b = []
  for j in p_find_all:
    b.append(j.text.strip().replace('\n', '').replace('\t', ''))
  performlist.append(b)
print(len(performlist))
preprocessing = []
print(performlist)
du_jeju = performlist[0:]
for i in performlist:
  if i[1] == "무대분기점검":
    du_jeju.remove(i)
for i in performlist:
  if  '무대주간점검' in i:
    print(i)
    du_jeju.remove(i)

```

```
data1 = du_jeju
print(data1)
for i in data1:
  i[0] = '문예회관'+i[0]
  i[1] = i[1].replace('<대관>','')
  i[2] = i[2].replace('일시:','')
print(data1)
Zdata = []
for j in data1:
  a = []
  a.append(j[1])
  a.append(j[0])
  a.append(j[2])
  a.append(j[3])
  Zdata.append(a)
print(Zdata)
```

```
seogipo_data = requests.get('http://culture.seogwipo.go.kr/artcenter/guide/schedule.htm?year=2020&month=10')
seogipo_data_soup = BeautifulSoup(seogipo_data.text,'html.parser')
ex_container2 = seogipo_data_soup.find_all('div',{'class':"ex-container"})
performlist_seogipo = []
for i in ex_container2:
  p_find_all = i.find_all('p')
  b = []
  for j in p_find_all:
    b.append(j.text.strip().replace('\n', '').replace('\t', ''))
  performlist_seogipo.append(b)
print((performlist_seogipo))
du_seogi = performlist_seogipo[0:]
for i in performlist_seogipo:
  if i[0] == '기타':
    du_seogi.remove(i)
print(du_seogi)
# 대관 세트 기획 전시
```

```
data2 = du_seogi
print(data2)
for i in data2:
  i[0] = '예술의전당'+i[0]
  i[1] = i[1].replace('대관','')
  i[1] = i[1].replace('세트','')
  i[1] = i[1].replace('기획','')
  i[1] = i[1].replace('전시','')
  i[2] = i[2].replace('일시:','')
print(data2)
Zdata2 = []
for j in data2:
  a = []
  a.append(j[1])
  a.append(j[0])
  a.append(j[2])
  a.append(j[3])
  Zdata2.append(a)
print(Zdata2)
```

```
artcenter_data = requests.get('https://www.jejusi.go.kr/acenter/show/perf.do?year=2020&month=10')
artcenter_data_soup = BeautifulSoup(artcenter_data.text,'html.parser')
info = artcenter_data_soup.find_all('ul',{'class':"info"})
performlist_art= []
for i in info:
  li_find_all = i.find_all('li')
  b = []
  for j in li_find_all:
    b.append(j.text.strip().replace('\n', '').replace('\t', ''))
  performlist_art.append(b)


du_art =   performlist_art[0:]


title = artcenter_data_soup.find_all('span',{'class':'title'})
for k in range(len(title)):
  du_art[k] = [title[k].text]+du_art[k]
print(du_art)
du_du_art = du_art[0:]
Zdata3 = []
for i in du_art:
  a = []
  a.append(i[0])
  a.append(i[3])
  a.append(i[2])
  a.append('')
  Zdata3.append(a)
print(Zdata3)

  
```

```
import pandas as pd
print(Zdata)
print(Zdata2)
print(Zdata3)


```

```
condata = Zdata+Zdata2+Zdata3 #Zdata 는 표준화된 리스트들이 들어 있는 리스트임.

dic = {'Title':[],'Place':[],'Date':[],"Time":[]}
for i in condata:
  dic['Title'].append(i[0])
  dic['Place'].append(i[1])
  dic['Date'].append(i[2])
  dic["Time"].append(i[3])
df = pd.DataFrame(dic)
```
